```python
import pandas as pd
file_path = '/content/drive/My Drive/project/alzheimers_disease_data.csv'  # Correct path
df = pd.read_csv(file_path)
df.head()
```

```python
feature = df.drop(['PatientID', 'DoctorInCharge'], axis=1)
feature.Diagnosis.value_counts()
```

```python
df.head().T
```

```python
#Before smoke
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
X_orig = feature.drop(columns=["Diagnosis"])
y_orig = feature["Diagnosis"]

X_train, X_test, y_train, y_test = train_test_split(X_orig, y_orig, test_size=0.2, random_state=42)

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))

```

```python
#After smoke
```

```python
from imblearn.over_sampling import SMOTE
X = feature.drop(columns=["Diagnosis"])
y = feature["Diagnosis"]
smote = SMOTE(random_state=42)
X_sm, y_sm = smote.fit_resample(X, y)

# Check new class distribution
print(y_sm.value_counts(normalize=True) * 100)
```

```python
print (y_sm.value_counts())
```

```python
x_train,x_test,y_train,y_test=train_test_split(X_sm,y_sm,test_size=0.2,random_state=42)
model = RandomForestClassifier(random_state=42)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

print(classification_report(y_test, y_pred))
```

```python
#Befoe smoke, the accuracy is 0.93. After smoke, the accuracy is 0.89.
#Overall accuracy is higher before smoke.
```
